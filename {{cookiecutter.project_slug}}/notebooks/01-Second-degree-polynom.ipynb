# Solve a second degree equation

Import some libraries and our `roots_second_degree` function, which can be found inside the `extra` folder.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

from extra import roots_second_degree
```

Let's define some random polynom:

$$ p(x) = -0.3 x^2 + 3.7 x - 5.1$$

by defining its coefficients:

```python
a, b, c = -0.3, 3.7, -5.1
```

And plot it

```python
x = np.linspace(-5, 15, 101)
y = a * x ** 2 + b * x + c

plt.plot(x, y)
plt.axhline(0, linestyle='--', c="grey")
plt.show()
```

Get the roots of the polynom

```python
roots = roots_second_degree(a, b, c)
roots
```

Add them to the plot and save the figure to `manuscript/figs`.

```python
plt.plot(x, y, label=r"$p(x) = -0.3 x^2 + 3.7 x - 5.1$")
plt.axhline(0, linestyle='--', c="grey")
plt.scatter(roots, [0, 0], c="C1", s=100, label="Roots of the polynom")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig(Path("..") / "manuscript" / "figs" / "second-degree-polynom-roots.pdf")
plt.show()
```
